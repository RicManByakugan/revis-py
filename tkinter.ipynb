```python
import tkinter as tk
import webbrowser

def open_url():
    webbrowser.open_new("https://www.google.com")

window = tk.Tk()
window.title("First App")
window.geometry("1080x700")
window.minsize(600, 600)
window.configure(background="#41B77E")
window.iconbitmap("logo.ico")

frame = tk.Frame(window, bg="#41B77E")
# frame = tk.Frame(window, bg="#41B77E", bd=1, relief="sunken")

label_title = tk.Label(frame, text="Welcome to my first app", font=("Arial", 30), bg="#41B77E", fg="white")
label_title.pack(expand="yes")
# label_title.pack(side="top")

label_subtitle = tk.Label(frame, text="SubTItle bla bla SubTItle bla bla", font=("Arial", 15), bg="#41B77E", fg="white")
label_subtitle.pack(expand="yes")

p_button = tk.Button(frame, text="Continuer", font=("Arial", 10), bg="white", fg="#41B77E", command=open_url, border=0)
p_button.pack(pady=25, expand="yes", fill="x")

frame.pack(expand="yes")

window.mainloop()
```

```python
from tkinter import *
import string
import random

def generate():
    pass_min = 6
    pass_max = 12
    all_char = string.ascii_letters + string.digits + string.punctuation
    password_characters = random.choices(all_char, k=random.randint(pass_min, pass_max))
    password = "".join(password_characters)
    entry_password.delete(0, END)
    entry_password.insert(0, password)
    
window = Tk()
window.title("Password Generator")
window.geometry("1080x700")
window.minsize(600, 600)
window.iconbitmap("logo.ico")
window.configure(background="#4065A4")

frame = Frame(window, bg="#4065A4")

width = 300
height = 300
image = PhotoImage(file="verrouillage.png").zoom(35).subsample(32)

canvas = Canvas(frame, width=width, height=height, bg="#4065A4", bd=0, highlightthickness=0)
canvas.create_image(width/2, height/2, image=image)
canvas.grid(row=0, column=0, sticky=W)

right_frame = Frame(frame, bg="#4065A4")

label_title = Label(right_frame, text="Password Generator", font=("Arial", 20), bg="#4065A4", fg="white")
label_title.pack()

entry_password = Entry(right_frame, font=("Arial", 15), bg="#4065A4", fg="white")
entry_password.pack(padx=20, pady=20)

button_generate = Button(right_frame, text="Générer", font=("Arial", 15), bg="#4065A4", fg="white", command=generate)
button_generate.pack(padx=20, pady=20, fill=X)

right_frame.grid(row=0, column=1, sticky=W)

frame.pack(expand=YES)


# menu
menu_bar = Menu(window)

file_menu = Menu(menu_bar, tearoff=0)
file_menu.add_command(label="New", command=generate)
file_menu.add_command(label="Quit", command=window.quit)

menu_bar.add_cascade(label="File", menu=file_menu)

edit_menu = Menu(menu_bar, tearoff=0)
edit_menu.add_command(label="Undo", command=generate)
edit_menu.add_command(label="Redo", command=generate)
menu_bar.add_cascade(label="Edit", menu=edit_menu)

window.config(menu=menu_bar)

window.mainloop()
```
